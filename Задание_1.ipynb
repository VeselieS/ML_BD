**Ответьте на следующие вопросы:**


1.   Какие средние значения в каждой числовой колонке вашего датасета?
2.   Минимальные и максимальные значения в каждой числовой колонке вашего датасета



```python
# Вынесем в разные списки названия колонок для числовых и нечисловых.
numericArr = []
otherArr = []

for i in df.columns:
    if df[i].dtype == "int64":
        numericArr.append(i)
    else:
        otherArr.append(i)

print("numericArr = ", numericArr)
print("otherArr = ", otherArr)
```

```python
# Ответ 1
print(df[numericArr].mean())
```

```python
# Ответ 2.1 (Min)
print(df[numericArr].min())
```

```python
# Ответ 2.2 (Max)
print(df[numericArr].max())
```

##Типизация
Присвойте соответствующим колонкам один из типов (category, float, int)

```python
for i in numericArr:
    df[i] = df[i].astype('Int64')

for j in otherArr:
    df[j] = df[j].astype('category')

df.dtypes
# df.info()
```
